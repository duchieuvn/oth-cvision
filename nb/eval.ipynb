```python
import json
import matplotlib.pyplot as plt


metrics_path = '../results/BUSI/train/UNetPlusPlusSum_metrics.json'
with open(metrics_path, 'r') as f:
    metrics = json.load(f)

# Extract each metric into a separate list
epochs = [entry['epoch'] for entry in metrics]
train_loss = [entry['train_loss'] for entry in metrics]
val_loss = [entry['val_loss'] for entry in metrics]
train_iou = [entry['train_iou'] for entry in metrics]
val_iou = [entry['val_iou'] for entry in metrics]
epoch_time = [entry['epoch_time'] for entry in metrics]

# Create subplots: Loss and IoU
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot Loss
axs[0].plot(epochs, train_loss, label='Train Loss', marker='o')
axs[0].plot(epochs, val_loss, label='Validation Loss', marker='x')
axs[0].set_title('Loss over Epochs')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
axs[0].legend()
axs[0].grid(True)

# Plot IoU
axs[1].plot(epochs, train_iou, label='Train IoU', marker='o')
axs[1].plot(epochs, val_iou, label='Validation IoU', marker='x')
axs[1].set_title('IoU over Epochs')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('IoU')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
